# 数据展示

```python
import pandas as pd 
hair_dryer = pd.read_csv('hair_dryer.tsv',sep='\t')
hair_dryer.head()
```

```python
review_body = pd.DataFrame(hair_dryer['review_body'])
review_body['review_body'].to_csv('./files/review_body.txt',index=None)
```

# 数据筛选与分析

```python
hair_dryer = hair_dryer[['product_id','product_parent','product_title','star_rating','helpful_votes','total_votes','vine','verified_purchase','review_headline','review_body','review_date']]
hair_dryer.drop(labels=(hair_dryer.loc[(hair_dryer['vine']=='N')&(hair_dryer['verified_purchase']=='N')].index),inplace=True)
hair_dryer.head()
```

```python
hair_dryer['vine'] = hair_dryer['vine'].replace('N',0)
hair_dryer['vine'] = hair_dryer['vine'].replace('Y',1)
hair_dryer['verified_purchase'] = hair_dryer['verified_purchase'].replace('N',0)
hair_dryer['verified_purchase'] = hair_dryer['verified_purchase'].replace('Y',1)
hair_dryer.to_csv('./files/hair_dryer.csv')
hair_dryer.head()
```

```python
print("Summary statistics of numerical features : \n", hair_dryer.describe()[['star_rating','helpful_votes','total_votes']])

print("\nTotal number of reviews: ",len(hair_dryer))
print("\nTotal number of vine=Y: ",len(hair_dryer[hair_dryer['vine']==1]))
print("\nTotal number of product: ", len(list(set(hair_dryer['product_id']))))

print("\nPercentage of reviews with neutral sentiment : {:.2f}%"\
      .format(hair_dryer[hair_dryer['star_rating']==3]["review_body"].count()/len(hair_dryer)*100))
print("\nPercentage of reviews with positive sentiment : {:.2f}%"\
      .format(hair_dryer[hair_dryer['star_rating']>3]["review_body"].count()/len(hair_dryer)*100))
print("\nPercentage of reviews with negative sentiment : {:.2f}%"\
      .format(hair_dryer[hair_dryer['star_rating']<3]["review_body"].count()/len(hair_dryer)*100))
```

# 数据可视化

```python
import matplotlib.pyplot as plt
vine_N = hair_dryer.star_rating[hair_dryer.vine == 0].value_counts()
vine_Y = hair_dryer.star_rating[hair_dryer.vine == 1].value_counts()
df = pd.DataFrame({'vine = N':vine_N,'vine = Y':vine_Y})
df.plot(kind = 'bar',stacked = True)
plt.title('Distribution of star_rating',fontsize=15)
plt.xlabel('star_rating',fontsize=15)
plt.ylabel('Count',fontsize=15)
plt.show()
```

```python
data = [8.80,78.68,12.52]
index = ['neutral','positive','negative']
df_data = pd.DataFrame(data=data,index=index)
plt.figure(figsize=(6,6))
plt.pie(df_data,labels=index,autopct='%1.2f%%')
plt.axis('equal')
plt.title('Percentage of positive,neutral and negative sentiments',fontsize=15)
```

```python
star_rating = hair_dryer.groupby(['star_rating']).count()['review_date']
star_rating
```

```python
plt.pie(star_rating,labels=['one star','two star','three star','four star','five star'],autopct='%1.2f%%')
plt.axis('equal')
plt.title('The proportion of each star_rating')
plt.show()
```

# 奶嘴同理

```python
pacifier = pd.read_csv('pacifier.tsv',sep='\t')
pacifier.head() 
pacifier = pacifier.dropna(axis=0,subset=['review_date'])
pacifier = pacifier[['product_id','product_parent','product_title','star_rating','helpful_votes','total_votes','vine','verified_purchase','review_headline','review_body','review_date']]
pacifier.drop(labels=(pacifier.loc[(pacifier['vine']=='N')&(pacifier['verified_purchase']=='N')].index),inplace=True)
pacifier['vine'] = pacifier['vine'].replace('N',0)
pacifier['vine'] = pacifier['vine'].replace('Y',1)
pacifier['vine'] = pacifier['vine'].replace('n',0)
pacifier['vine'] = pacifier['vine'].replace('y',1)
pacifier['verified_purchase'] = pacifier['verified_purchase'].replace('N',0)
pacifier['verified_purchase'] = pacifier['verified_purchase'].replace('Y',1)
pacifier['verified_purchase'] = pacifier['verified_purchase'].replace('n',0)
pacifier['verified_purchase'] = pacifier['verified_purchase'].replace('y',1)
pacifier.to_csv('./files/pacifier.csv')
pacifier.head()
```

```python
screening_pacifier = pacifier[['product_id','product_parent','product_title','star_rating','helpful_votes','total_votes','vine','verified_purchase','review_headline','review_body','review_date']]
review_data = screening_pacifier['review_headline'] +  ' ' + screening_pacifier['review_body']
pacifier1 = pacifier[['star_rating','helpful_votes','total_votes']]
pacifier1['review'] = review_data
pacifier1.to_csv('./files/pacifier1.csv')
clean_review = review_data.str.replace('!','')
clean_review = clean_review.str.replace('<br />',' ')
clean_review = clean_review.str.replace('.','')
clean_review = clean_review.str.replace(';','')
clean_review = clean_review.str.replace(':','')
clean_review = clean_review.str.replace('-','')
clean_review = clean_review.str.replace(',','')
clean_review = clean_review.str.replace('*','')
clean_review = clean_review.str.replace('#','')
clean_review = clean_review.str.replace('&','')
clean_review = clean_review.str.replace('(','')
clean_review = clean_review.str.replace(')','')
clean_review = clean_review.str.replace('/','')
clean_review = clean_review.str.replace('  ',' ')
screening_pacifier['review'] = clean_review
screening_pacifier.to_csv('./files/screening_pacifier.csv')
screening_pacifier.head()
```

```python
df_clean_review = pd.DataFrame(clean_review.astype(str))
df_clean_review.to_csv('./files/pacifier_clean_review.txt',index=None,header=None)
df_clean_review.head()
import jieba
mycut = lambda s: ' '.join(jieba.cut(s))
cut_review = df_clean_review[0].apply(mycut)
cut_review.to_csv('./files/pacifier_cut_review.txt',index = False, header = False)
cut_review
```

```python
stop = pd.read_csv('stoplist_pacifier.txt',header = None)
stop = [' ', ''] + list(stop[0])
cut_review1 = cut_review.apply(lambda s: s.split(' ')) #定义一个分割函数，然后用apply广播,每行按空格分隔,每行由str转变为list
cut_review2 = cut_review1.apply(lambda x: [i for i in x if i not in stop]) #逐词判断是否停用词
# cut_review2
from gensim import corpora,models
review_dict = corpora.Dictionary(cut_review2) #建立词典
review_corpus = [review_dict.doc2bow(i) for i in cut_review2] #建立语料库
review_lda = models.LdaModel(review_corpus, num_topics = 3, id2word = review_dict,passes=2) #LDA模型训练
print('review:')
for review_info in review_lda.show_topics():
  print(review_info)#展示主题
```
