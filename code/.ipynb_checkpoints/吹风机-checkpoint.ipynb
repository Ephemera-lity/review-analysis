# 余弦相似

```python
import pandas as pd 
import numpy as np
import jieba
hair_dryer = pd.read_csv('./files/hair_dryer1.csv',encoding='gb18030')
hair_dryer = hair_dryer[['star_rating','review']]
hair_dryer.head()
#进行分词
mycut = lambda s: ' '.join(jieba.cut(s))
cut_review = hair_dryer['review'].astype(str).apply(mycut)
review_2000 = cut_review[0:2000]
review_2000.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer, TfidfTransformer
from sklearn.metrics.pairwise import cosine_similarity
# 语料库
corpus = list(review_2000[::1])
# TF-IDF向量化工具
tfidf_vec = TfidfVectorizer() 
tfidf_matrix = tfidf_vec.fit_transform(corpus)
# 计算各评论的相似度
cosine_similarity = cosine_similarity(tfidf_matrix)
#吹风机第特定条评论跟前1957条评论的相似度 94-10days  361-31days  1957-175days
df_cosine_similarity = pd.DataFrame(cosine_similarity[[8,11,56,58,61,62,78,1027]])
df_cosine_similarity = df_cosine_similarity.iloc[:,0:1957]
df_cosine_similarity
```

```python
list1 = []
for i in range(0,1957):
    max_num =max(df_cosine_similarity[i])
    j=0
    for each_num in df_cosine_similarity[i]:
        if max_num > 0.0:
            if each_num == max_num:
                list1.append(j)
            j = j+1
        else:
            list1.append(int(max_num))
            break
print(list1)
```

```python
#对应类别所在索引
review_kind1 = []
for j in list1:
    if j == 0:
        review_kind1.append(11)
    elif j == 1:
        review_kind1.append(14)
    elif j == 2:
        review_kind1.append(62)
    elif j == 3:
        review_kind1.append(64)
    elif j == 4:
        review_kind1.append(67)
    elif j == 5:
        review_kind1.append(68)
    elif j == 6:
        review_kind1.append(85)
    else:
        review_kind1.append(1071)
print(review_kind1)
```

```python
#对应类别的分数
review_score1 = []
for j in list1:
    if j == 0:
        review_score1.append(6)
    elif j == 1:
        review_score1.append(-2)
    elif j == 2:
        review_score1.append(9)
    elif j == 3:
        review_score1.append(3)
    elif j == 4:
        review_score1.append(1)
    elif j == 5:
        review_score1.append(5)
    elif j == 6:
        review_score1.append(6)
    else:
        review_score1.append(-4)
print(review_score1)
```

```python
hair_dryer100_kind = hair_dryer[0:1957]
hair_dryer100_kind['review_kind'] = review_kind1
hair_dryer100_kind['review_score'] = review_score1
#hair_dryer100_kind.to_csv('hair_dryer100_kind.csv')
hair_dryer100_kind
```

# EWMA

```python
import matplotlib.pyplot as plt
hair_dryer1 = pd.read_csv('./files/hair_dryer.csv',encoding='gb18030')
hair_dryer100_kind['review_date'] = hair_dryer1['review_date']
sum_score = hair_dryer100_kind.groupby('review_date').sum()[['star_rating','review_score']]
#重新排列日期索引
sum_score['date'] = pd.to_datetime(sum_score.index).values
sum_score.set_index('date',inplace=True)
sum_score.sort_index(inplace=True)
total_score = sum_score['star_rating'] * 0.3 + sum_score['review_score'] * 0.7
sum_score['total_score'] = total_score
sum_score.head()
```

```python
sum_score = sum_score.loc['2015-08-01':'2015-08-31']
def EWMA(scores, alpha=0.9):
    newScores = np.zeros(shape = scores.shape)
    newScores[0] = scores[0]
    for i in range(1, len(scores)):
        newScores[i] = scores[i]*alpha + newScores[i-1]*(1-alpha)
    return newScores;
import pandas as pd 
scores = sum_score['total_score']
newScores = EWMA(scores, alpha=0.1)
sum_score['EWMAstar_rating'] = newScores
print(sum_score.head())
fig, ax = plt.subplots(figsize=(8,6))
plt.plot(sum_score.index,sum_score['EWMAstar_rating'],marker='*')
plt.xlabel('Date',fontsize=15)
plt.ylabel('EWMAScores',fontsize=15)
plt.title('EWMAScores of hair_dryer',fontsize=15)
plt.legend(['hair_dryer'],fontsize=15)
```

```python
hair_dryer1 = pd.read_csv('./files/hair_dryer.csv',encoding='gb18030')
hair_dryer100_kind['review_date'] = hair_dryer1['review_date']
#求每天平均的星级、评论分和综合评分
ave_score = hair_dryer100_kind.groupby('review_date').mean()[['star_rating','review_score']]
#排列日期
ave_score['date'] = pd.to_datetime(ave_score.index).values
ave_score.set_index('date',inplace=True)
ave_score.sort_index(inplace=True)
print(ave_score.head())
ave_total_score = ave_score['star_rating'] * 0.3 + ave_score['review_score'] * 0.7
ave_score['ave_total_score'] = ave_total_score
ave_score
```

```python
ave_score2 = ave_score.loc['2015-08-01':'2015-08-31']
# plt.plot(ave_score2.index,ave_score2['star_rating'],marker='*')
def EWMA(scores, alpha=0.9):
    newScores = np.zeros(shape = scores.shape)
    newScores[0] = scores[0]
    for i in range(1, len(scores)):
        newScores[i] = scores[i]*alpha + newScores[i-1]*(1-alpha)
    return newScores;
import pandas as pd 
scores = np.array(ave_score2['star_rating'])
newScores = EWMA(scores, alpha=0.3)
ave_score2['EWMAstar_rating'] = newScores
# print(ave_score2.head())
fig, ax = plt.subplots(figsize=(8,6))
plt.plot(ave_score2.index,ave_score2['EWMAstar_rating'],marker='*')
plt.xlabel('Date',fontsize=15)
plt.ylabel('EWMAScores',fontsize=15)
plt.title('The average star_rating EWMAScores per day in one month',fontsize=15)
plt.legend(['hair_dryer'],fontsize=15)
```

# 用平均累加求和代入logistics函数求概率值判断是否潜在成功

```python
cumsum = 0
for i in range(0,30):
    cumsum1 = sum_score['total_score'][i+1] - sum_score['total_score'][i]
    cumsum = cumsum + cumsum1
print('cumsum=',cumsum/30)
# Logistic函数，输出0-1内的数，映射成概率值
def sigmoid(z):
    return 1.0/(1 + np.exp(-z))
print('吹风机概率值为',sigmoid(cumsum/30))
```

# Pearson相关系数

```python
import numpy as np
import pandas as pd
a= np.array([2,3,9,6,8])
b= np.array([5,6,3,7,9])
c=np.array([a,b])
#axis=0,表示按列求均值
c_mean= np.mean(c,axis=0) 
c_std = np.std(c,axis=0)
c_zscore= (c-c_mean)/c_std #标准化
# c_zscore_corr= np.corrcoef(c_zscore) #相关系数矩阵
# print(c_zscore_corr)
c_pd = pd.DataFrame(c_zscore.T, columns=['c1', 'c2'])
c_corr= c_pd.corr(method='spearman') 
c_corr
```
